```python
import numpy as np

def spiral(n):
    matrix = np.zeros((n, n), dtype=int)
    row, col = n // 2, n // 2
    num = 1
    step = 1
    
    matrix[row, col] = num

    while step < n:
        # Move to the right
        for _ in range(step):
            col += 1
            num += 1
            matrix[row, col] = num

        # Move up
        for _ in range(step):
            row -= 1
            num += 1
            matrix[row, col] = num

        # Increase step size
        step += 1

        # Move to the left
        for _ in range(step):
            col -= 1
            num += 1
            matrix[row, col] = num

        # Move down
        for _ in range(step):
            row += 1
            num += 1
            matrix[row, col] = num

        # Increase step size
        step += 1

    # Fill the last row
    for _ in range(step - 1):
        col += 1
        num += 1
        matrix[row, col] = num

    return matrix

# Example usage:
n = 3
spiral_matrix = spiral(n)
print("Spiral for n = ", n,":")
print()
print(spiral_matrix)
print()

n = 5
spiral_matrix = spiral(n)
print("Spiral for n = ", n,":")
print()
print(spiral_matrix)
print()

n = 7
spiral_matrix = spiral(n)
print("Spiral for n = ", n,":")
print()
print(spiral_matrix)
print()

n = 15
spiral_matrix = spiral(n)
print("Spiral for n = ", n,":")
print()
print(spiral_matrix)
print()
```
