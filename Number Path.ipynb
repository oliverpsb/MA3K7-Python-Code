```python
import random
import matplotlib.pyplot as plt

def game(N):
    hits = 0
    misses = 0
    for i in range(N):
        counter = 1
        while counter <= 30:
            counter += random.randint(1,2)
            if counter == 25:
                hits += 1
                break
            elif counter > 25:
                misses +=1
                break
    ratio = hits/(hits+misses)
    print("The proportion of hits is: ",ratio)
    return
game(100000)
```
