```python
import random

class Particle:
    def __init__(self, species):
        self.species = species

def collide(p1, p2):
    if p1.species != p2.species:
        remaining_species = {'A', 'B', 'C'} - {p1.species, p2.species}
        new_species = remaining_species.pop()
        p1.species = new_species
        p2.species = new_species

def game(T):
    # Initialize particles with at least one of each species
    particles = [Particle('A'), Particle('B'), Particle('C')]
    
    # Add additional particles
    for _ in range(T - 3):
        particles.append(Particle(random.choice(['A', 'B', 'C'])))
    
    initial = {species: sum(1 for p in particles if p.species == species) for species in ['A', 'B', 'C']}
    print("Initial Species Count:")
    print(initial)
    
 
    
    while len(set(p.species for p in particles)) > 1:
        particle1, particle2 = random.sample(particles, k=2)
        collide(particle1, particle2)
    
    final = particles[0].species
    
    print("Final Species Count: ",final)

T =  10
#Looking at a random game for when T = 10
game(T)
```
